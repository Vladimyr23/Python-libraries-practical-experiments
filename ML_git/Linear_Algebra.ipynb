# Exercise 6: The inverse of a matrix

Firstly, compute Matrix AA, where
$$AA = A\times A^T$$

Secondly, compute the inverse of Matrix $AA$

**Hint:** you can use linalg class of numpy to compute the inverse of a matrix

```python
import numpy.linalg as linalg

# A * A^T ... we can only invert quadratic matrices
AA = A.dot(A.transpose()) 

# the inverse of Matrix AA
AAinv = linalg.inv(AA)

AA, AAinv
```

Thirdly, multiplying Matrix $AA$ with its inverse $(AA)^{-1}$ needs to result in the Identity matrix. That is,
$$AA \times (AA)^{-1} = I$$

Fourthly, print your results.

```python
# AA * AA^(-1)
AA.dot(AAinv), AAinv.dot(AA) 
```

**Note:** Computing the inverse of a matrix is tricky and it is hard to get a numerically accurate solution. Whenever we need to compute the inverse of a matrix times another matrix $\boldsymbol({AA})^{-1}\boldsymbol{h}$, then it is better to use specifically tailored methods for this which are numerically more stable. 

```python
import numpy.linalg as linalg
# compute AA^-1*h in a more stable way using linalg.solve.
h = np.array([1, 2])
out1 = linalg.solve(AA, h)

out1
```

## Exercise 7: More exercises for vector and matrix calculations
Compute: 
* $(\boldsymbol{A} \boldsymbol a - \boldsymbol b)^T(\boldsymbol A \boldsymbol a - \boldsymbol b)$, 
* $(\boldsymbol{C} \boldsymbol b)^T\boldsymbol C$
* $(\boldsymbol{C}^T \boldsymbol C)^{-1}\boldsymbol C^T \boldsymbol a$, 

where
$$ A = \left( \begin{array}{cc} 1 & 0 & 1 \\ 2 & 3 & 1 \end{array}\right)$$

$$ C = \left( \begin{array}{cc} 1 & 0 \\ 2 & 3 \\ 1 & 5 \end{array}\right)$$

$$ a = \left( \begin{array}{cc} 1 & 2 & 1 \end{array}\right)$$

$$ b = \left( \begin{array}{cc} 2 & 2 \end{array}\right)$$

Check your result also in terms of the dimensionalities of the resulting matrices. That is an easy way of spotting an error. Always use the linalg.solve method instead of the linalg.inv method if possible. 

```python
# Create Matrix A, C
A = np.array([[1, 0, 1], [2, 3, 1]])
C = np.array([[1, 0], [2, 3], [1, 5]])

# Create Vector a, b
a = np.array([1,2,1])
b = np.array([2,2])

#  Caculate (Aa-b)^T * (Aa -b)
temp = A.dot(a) - b
sol1 = temp.transpose().dot(temp)

# Caculate (Cb)^T * C
temp = C.dot(b)
sol2 = temp.transpose().dot(C)

# Caculate (C^T * C)^(-1) * C^T * a
temp = C.transpose().dot(C)
sol3 = linalg.solve(temp, C.transpose().dot(a))

# Print results of three above equations
sol1, sol2, sol3
```

## Exercise 8: Singular Value Decomposition (SVD)

Conduct SVD on Matrix $G$ and obtain Matrices U, D, and V, where

$$ G = \left( \begin{array}{cc} 7 & 2 \\ 3 & 4 \\ 5 & 3 \end{array}\right)$$

**Hint:** The singular value decomposition can be done with the linalg.svd() function from Numpy 

```python
# Create Matrix G
G = np.array([[7, 2], [3, 4], [5, 3]])

# Compute Matrices U, D and V
U, D, V = np.linalg.svd(G)

U, D, V
```

Show the dimensions of Matrices U, D, and V

Letâ€™s start by loading an image in python and convert it to a Numpy array. We will convert it to grayscale to have one dimension per pixel. The shape of the matrix corresponds to the dimension of the image filled with intensity values: 1 cell per pixel.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load "test_svd.png" image
plt.style.use('classic')
img = Image.open('test_svd.png')
# convert image to grayscale
imggray = img.convert('LA')

# convert to numpy array
imgmat = np.array(list(imggray.getdata(band=0)), float)
# Reshape according to orginal image dimensions
imgmat.shape = (imggray.size[1], imggray.size[0])

plt.figure(figsize=(9, 6))
plt.imshow(imgmat, cmap='gray')
plt.show()
```

Compute Matrix $U$, $D$, and $V$ by Singular Value Decomposition (SVD).

**Hint:** You can use np.linalg.svd() function to obtain Matrix $U$, $D$, and $V$

```python
# Compute Matrix U, D, and V
U, D, V = np.linalg.svd(imgmat)
```

Reconstruct the image from two singular values.

```python
# Provide your code here
reconstimg = np.matrix(U[:, :2]) * np.diag(D[:2]) * np.matrix(V[:2, :])
plt.imshow(reconstimg, cmap='gray')
plt.show()
```

Draw the reconstructed image using different number of singular values, e.g. 5, 10, 15, 20, 30, and 50.

```python
# reconstructed image with different number of singular values
for i in [5, 10, 15, 20, 30, 50]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
